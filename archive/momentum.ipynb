# Functions / Classes

```python
class TickerObj:
    def __init__(self, name):
        self.ticker = name
        self.obj = yf.Ticker(name)
        self.data = self.obj.history(start=DT_3YR_PRIOR, end=DT_STATIC)
        self.info = self.obj.info
```

# Import Data

```python
df_spy = TickerObj('SPY').data
```

```python
df_spy['DailyPercentageChange'] = (df_spy['Close'] - df_spy.shift(1)['Close'])/df_spy.shift(1)['Close']
```

```python
for i in [5, 10, 20, 50]:
    df_spy[f'Ma{i}Day'] = df_spy.shift(1)['Close'].rolling(i).mean()
```
